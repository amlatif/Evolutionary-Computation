```python
# Objective function
def f(x):
    return x ** 2 + 5 * x + 6
```

```python
# PSO parameters
num_particles = 30
max_iter = 100
w = 0.5       # Inertia weight
c1 = 1.5      # Cognitive coefficient
c2 = 1.5      # Social coefficient
```

```python
# Initialize particles
particles = np.random.uniform(-10, 10, num_particles)
velocities = np.random.uniform(-1, 1, num_particles)
```

```python
# Initialize personal best and global best
pbest = particles.copy()
pbest_fitness = np.array([f(x) for x in particles])
gbest = pbest[np.argmin(pbest_fitness)]
gbest_fitness = min(pbest_fitness)
```

```python
# PSO loop
for iteration in range(max_iter):
    for i in range(num_particles):
        # Update velocity
        r1, r2 = np.random.rand(), np.random.rand()
        velocities[i] = (w * velocities[i] +
                         c1 * r1 * (pbest[i] - particles[i]) +
                         c2 * r2 * (gbest - particles[i]))
        
        # Update position
        particles[i] += velocities[i]
        
        # Ensure position stays within bounds
        particles[i] = np.clip(particles[i], -10, 10)
        
        # Evaluate fitness
        fitness = f(particles[i])
        
        # Update personal best
        if fitness < pbest_fitness[i]:
            pbest[i] = particles[i]
            pbest_fitness[i] = fitness
        
        # Update global best
        if fitness < gbest_fitness:
            gbest = particles[i]
            gbest_fitness = fitness
    
    print(f"Iteration {iteration+1}: Best Fitness = {gbest_fitness}")

print(f"Optimal Solution: x = {gbest}, f(x) = {gbest_fitness}")
```

```python
# تابع هدف
def fitness(grid):
    penalty = 0
    
    # بررسی ردیف‌ها و ستون‌ها
    for i in range(9):
        row_counts = np.bincount(grid[i, :], minlength=10)[1:]
        col_counts = np.bincount(grid[:, i], minlength=10)[1:]
        penalty += np.sum(row_counts[row_counts > 1])
        penalty += np.sum(col_counts[col_counts > 1])
    
    # بررسی بلوک‌های $3 \times 3$
    for block_row in range(3):
        for block_col in range(3):
            block = grid[block_row*3:(block_row+1)*3, block_col*3:(block_col+1)*3]
            block_counts = np.bincount(block.ravel(), minlength=10)[1:]
            penalty += np.sum(block_counts[block_counts > 1])
    
    return penalty
```

```python
import numpy as np

# مثال جدول سودوکو (0 نشان‌دهنده خانه خالی است)
sudoku = np.array([
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
])

# تابع هدف
def fitness(grid):
    penalty = 0
    
    # بررسی ردیف‌ها و ستون‌ها
    for i in range(9):
        row_counts = np.bincount(grid[i, :], minlength=10)[1:]
        col_counts = np.bincount(grid[:, i], minlength=10)[1:]
        penalty += np.sum(row_counts[row_counts > 1])  # جریمه برای تکرار در ردیف
        penalty += np.sum(col_counts[col_counts > 1])  # جریمه برای تکرار در ستون
    
    # بررسی بلوک‌های $3 \times 3$
    for block_row in range(3):
        for block_col in range(3):
            block = grid[block_row*3:(block_row+1)*3, block_col*3:(block_col+1)*3]
            block_counts = np.bincount(block.ravel(), minlength=10)[1:]
            penalty += np.sum(block_counts[block_counts > 1])  # جریمه برای تکرار در بلوک
    
    return penalty

# پارامترهای PSO
num_particles = 50
max_iter = 1000
w_max = 0.9  # حداکثر وزن اینرسی
w_min = 0.4  # حداقل وزن اینرسی
c1 = 1.5     # ضریب شناختی
c2 = 1.5     # ضریب اجتماعی

# مقداردهی اولیه ذرات
particles = []
for _ in range(num_particles):
    particle = sudoku.copy()
    for r in range(9):
        empty_indices = np.where(particle[r] == 0)[0]
        if len(empty_indices) > 0:
            values = np.random.permutation(np.setdiff1d(range(1, 10), particle[r]))[:len(empty_indices)]
            particle[r, empty_indices] = values
    particles.append(particle)

pbest = particles.copy()
pbest_fitness = [fitness(p) for p in particles]
gbest = particles[np.argmin(pbest_fitness)]
gbest_fitness = min(pbest_fitness)

# الگوریتم PSO
for iteration in range(max_iter):
    w = w_max - (w_max - w_min) * (iteration / max_iter)  # کاهش وزن اینرسی
    for i in range(num_particles):
        new_particle = particles[i].copy()
        
        # جابجایی مقادیر در ردیف‌ها
        for r in range(9):
            empty_indices = np.where(sudoku[r] == 0)[0]
            if len(empty_indices) > 1:
                idx1, idx2 = np.random.choice(empty_indices, 2, replace=False)
                new_particle[r, [idx1, idx2]] = new_particle[r, [idx2, idx1]]
        
        # ارزیابی ذره جدید
        new_fitness = fitness(new_particle)
        if new_fitness < pbest_fitness[i]:
            pbest[i] = new_particle
            pbest_fitness[i] = new_fitness
        
        if new_fitness < gbest_fitness:
            gbest = new_particle
            gbest_fitness = new_fitness
    
    print(f"Iteration {iteration+1}: Best Fitness = {gbest_fitness}")
    
    if gbest_fitness == 0:
        break

print("Solved Sudoku:")
print(gbest)
```

```python
import numpy as np

# مثال جدول سودوکو (0 نشان‌دهنده خانه خالی است)
sudoku = np.array([
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
])

# تابع هدف
def fitness(grid):
    penalty = 0
    
    # بررسی ردیف‌ها و ستون‌ها
    for i in range(9):
        row_counts = np.bincount(grid[i, :], minlength=10)[1:]
        col_counts = np.bincount(grid[:, i], minlength=10)[1:]
        penalty += np.sum(row_counts[row_counts > 1])  # جریمه برای تکرار در ردیف
        penalty += np.sum(col_counts[col_counts > 1])  # جریمه برای تکرار در ستون
    
    # بررسی بلوک‌های $3 \times 3$
    for block_row in range(3):
        for block_col in range(3):
            block = grid[block_row*3:(block_row+1)*3, block_col*3:(block_col+1)*3]
            block_counts = np.bincount(block.ravel(), minlength=10)[1:]
            penalty += np.sum(block_counts[block_counts > 1])  # جریمه برای تکرار در بلوک
    
    return penalty

# پارامeterهای PSO
num_particles = 50
max_iter = 1000
c1 = 1.5     # ضریب شناختی
c2 = 1.5     # ضریب اجتماعی

# مقداردهی اولیه ذرات
particles = []
for _ in range(num_particles):
    particle = sudoku.copy()
    for r in range(9):
        empty_indices = np.where(particle[r] == 0)[0]
        if len(empty_indices) > 0:
            values = np.random.permutation(np.setdiff1d(range(1, 10), particle[r]))[:len(empty_indices)]
            particle[r, empty_indices] = values
    particles.append(particle)

pbest = particles.copy()
pbest_fitness = [fitness(p) for p in particles]
gbest = particles[np.argmin(pbest_fitness)]
gbest_fitness = min(pbest_fitness)

# الگوریتم PSO
for iteration in range(max_iter):
    for i in range(num_particles):
        new_particle = particles[i].copy()
        
        # عملگر Crossover (ترکیب با pbest و gbest)
        for r in range(9):
            empty_indices = np.where(sudoku[r] == 0)[0]
            if len(empty_indices) > 0:
                # ترکیب با pbest
                if np.random.rand() < c1 / (c1 + c2):
                    new_particle[r, empty_indices] = pbest[i][r, empty_indices]
                # ترکیب با gbest
                else:
                    new_particle[r, empty_indices] = gbest[r, empty_indices]
        
        # عملگر Mutation (جهش تصادفی)
        for r in range(9):
            empty_indices = np.where(sudoku[r] == 0)[0]
            if len(empty_indices) > 0 and np.random.rand() < 0.1:  # احتمال جهش
                idx = np.random.choice(empty_indices)
                new_particle[r, idx] = np.random.choice(np.setdiff1d(range(1, 10), new_particle[r]))
        
        # ارزیابی ذره جدید
        new_fitness = fitness(new_particle)
        if new_fitness < pbest_fitness[i]:
            pbest[i] = new_particle
            pbest_fitness[i] = new_fitness
        
        if new_fitness < gbest_fitness:
            gbest = new_particle
            gbest_fitness = new_fitness
    
    print(f"Iteration {iteration+1}: Best Fitness = {gbest_fitness}")
    
    if gbest_fitness == 0:
        break

print("Solved Sudoku:")
print(gbest)
```
